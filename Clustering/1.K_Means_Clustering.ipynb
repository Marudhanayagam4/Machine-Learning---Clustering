```python
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv('Mall_Customers.csv')  # Load the dataset
dataset
```

```python
# Select relevant features 
X = dataset.iloc[:,3:5].values
X
```

```python
from sklearn.cluster import KMeans
list_1=[]
for i in range(1,11) :
    k_means=KMeans(n_clusters=i, init='k-means++', random_state=42)
    k_means.fit(X)
    list_1.append(k_means.inertia_)

# Elbow Method Plot
plt.plot(range(1,11),list_1)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = k_means.fit_predict(X) # Fit the model
```

```python
y_kmeans
```

```python
supervised=dataset
supervised
```

```python
supervised['Cluster_groups']=y_kmeans # Add the cluster labels to the original dataset
supervised
```

```python
#  Cluster Visualization
import seaborn as sns
facet=sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], fit_reg=False)
```
