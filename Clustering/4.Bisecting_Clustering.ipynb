```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X=dataset.iloc[:,3:5].values
X
```

```python
# Bisecting KMeans Clustering 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
inertias=[]
for k in range(1,11) :
    kmeans=KMeans(n_clusters=k, random_state=42).fit(X)
    inertias.append(kmeans.inertia_)
plt.plot(range(1,11),inertias)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertias')
plt.title('Elbow Method For Optimal k (Bisecting K-Means)')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import BisectingKMeans
bisect_means = BisectingKMeans(n_clusters=5, random_state=0)
```

```python
# Fit model
y_bisect_kmeans = bisect_means.fit_predict(X)
y_bisect_kmeans
```

```python
supervised=dataset
supervised
```

```python
supervised["Cluster_group"]=y_bisect_kmeans
supervised
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,
                x=supervised.columns[3],
                y=supervised.columns[4],
                hue=supervised.columns[5],
                fit_reg=False)
plt.title("Bisecting K-Means Clustering")
plt.show()
```
