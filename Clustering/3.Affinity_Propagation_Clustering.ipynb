```python
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv('Mall_Customers.csv')  # Load the dataset
dataset
```

```python
# Select relevant features
X=dataset.iloc[:,3:5].values
X
```

```python
from sklearn.cluster import AffinityPropagation
aff=AffinityPropagation(random_state=42) # don't need to specify the number of clusters in advance.
y_ap=aff.fit_predict(X)
```

```python
y_ap
```

```python
supervised=dataset
supervised
```

```python
supervised['Cluster_groups']=y_ap
supervised
```

```python
# Clustering Visualization
import seaborn as sns
facet=sns.lmplot(data=supervised, 
                x=supervised.columns[3],
                y=supervised.columns[4],
                hue=supervised.columns[5],
                fit_reg=False)
plt.title("Affinity Propagation Clustering")
plt.show()
```
